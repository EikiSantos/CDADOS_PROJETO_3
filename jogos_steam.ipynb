# Análise das avaliações de jogos na plataforma Steam

Neste experimento tentaremos fazer predições da positividade de avaliações dos jogos da plataforma Steam, utilizando regressões lineares. Utilizaremos os modelos de Regressão Linear Simples, Random Forest Regression e Regressão polinomial. Esse tipo de análise é relevante, pois através dele, como desenvolvedores de jogos, podemos tentar prever com base em características (como preço, número de categorias, tempo de lançamento, média/mediana de horas jogadas, entre outras), a recepção do produto pela comunidade.

Primeiramente faremos uma análise descritiva dos dados, adicionaremos modificações para melhor utilização nos modelos e, por fim verificaremos algumas correlações entre os dados apresentados pelo Dataset a seguir: https://www.kaggle.com/nikdavis/steam-store-games.

A partir destas análises, partiremos para a implementação nos modelos de regressão.

```python
#importando as bibliotecas necessárias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
#lendo o dataset e transformando em um dataframe
excel_jogos = pd.read_csv("steam-store-games/steam.csv")
```

# Generos existentes no dataset

```python
generos = excel_jogos.genres.str.cat(sep=";")
generos = generos.split(";")
lista_generos = []
for i in generos:
    if i not in lista_generos:
        lista_generos.append(i)
lista_software = ["Animation & Modeling", "Web Publishing", "Design & Illustration", "Early Access", "Utilities", "Video Production", "Nudity", "Photo Editing", "Accounting", "Game Development", "Free to Play", "Audio Production", "Tutorial", "Sexual Content", "Education", "Software Training", "Documentary"]
for i in lista_software:
    if i in lista_generos:
        lista_generos.remove(i)
lista_generos
```

Foi pesquisado sobre os generos existentes para remover os que não eram relacionadas a jogos, como por exemplo "Web Publishing", a fim de focar apenas nos jogos.

# DF só com jogos

```python
for i in lista_software:
    excel_only_jogos = excel_jogos[~excel_jogos.genres.str.contains(i)]
```

Foi criado um novo DataFrame a partir da lista de jogos obtida.

# Porcentagem dos feedbacks positivos e negativos

Foi feito a porcentagem dos feedbacks para evitar que jogos lançados a muito tempo tenham maior número de feedback, tanto positivo quanto negativo, dessa forma, a porcentagem considera apenas a relação entre os positivos, negativos e o total.

```python
excel_only_jogos["ratings"] = excel_only_jogos["positive_ratings"] + excel_only_jogos["negative_ratings"]
excel_only_jogos["Percent_positive"] = excel_only_jogos["positive_ratings"]/excel_only_jogos["ratings"]
excel_only_jogos["Percent_negative"] = excel_only_jogos["negative_ratings"]/excel_only_jogos["ratings"]
```

# Data de lançamento em anos

No dataset, a data de lançamento se dava por uma string contendo o ano, mês e dia, como por exemplo "2001-02-04", dessa forma, transformamos essa string em um número que seria à quanto tempo o jogo foi lançado

```python
l1 = []
l2 = []
l3 = []

for string in excel_only_jogos["release_date"]:
    l1.append(string.split("-"))

for lista in l1:
    l2.append(int(lista[0]))

for e in l2:
    l3.append(2019 - e)
excel_only_jogos["Idade"] = l3
```

# Número de categorias

Foi calculado o número de categorias para analisar jogos que podem ser considerados mais "Completos" por conter mais categorias e jogos mais "Simples" por ter poucas categorias.

```python
n_cat = []
for i in excel_only_jogos.categories:
    n_cat.append(i.count(";")+1)
excel_only_jogos["n_cat"] = pd.Series(n_cat)
```

# Excluindo jogos que não foram jogados (horas médias = 0)

```python
excel_only_jogos = excel_only_jogos.loc[excel_only_jogos.average_playtime != 0]
```

Para que possamos analisar os dados de maneira mais concreta, eliminamos os jogos que não foram jogados(jogos com tempo médio de jogo de 0), já que esses, por não terem sido jogados, não possuem ou possuem informações não confiáveis quanto às suas avaliações.

# Tempo médio por gênero

Para melhor observar como o tempo médio de jogo se relaciona com o gênero dele, foi criado um gráfico boxplot.

```python
t_medio = excel_only_jogos
data = [t_medio.loc[(t_medio.genres.str.contains(i)) & (t_medio.average_playtime > 0)].average_playtime for i in lista_generos]
fig, ax = plt.subplots(figsize=(20,15))
ax.set_title('Tempo médio de jogo por genêro')
ax.boxplot(data, labels = lista_generos)
plt.show()
```

# Log tempo médio por gênero

Como vemos acima, o gráfico boxplot de tempo médio por genêro não nos ajuda muito a perceber alguma relação real entre tempo médio e genêro. Portanto, também foi criado um boxplot de log de tempo médio de jogo por genêro para tentarmos ver se essa correlação existe mais facilmente.

```python
t_medio["log_average_playtime"] = np.log(t_medio["average_playtime"].loc[t_medio["average_playtime"]>0])
data = [t_medio.loc[(t_medio.genres.str.contains(i)) & (t_medio.average_playtime > 0)].log_average_playtime for i in lista_generos]
fig, ax = plt.subplots(figsize=(20,15))
ax.set_title('Tempo médio de jogo (LOG) por genêro')
ax.boxplot(data, labels = lista_generos)
plt.show()
```

Percebe-se que, utilizando a função Log, foi encontrado um "melhor" boxplot, porém não houve relação entre os generos e o tempo

De acordo com as correlações apresentadas, podemos afirmar que, de maneira geral, obtém-se regressões que possuem um resultado pobre, sem muito valor.
```python
reg = LinearRegression().fit(X_train, y_train)
```

```python
reg.coef_
print('Coeficientes estimados da regressão linear: {0}'.format(reg.coef_))
```

```python
reg.intercept_
print('Termo independente na regressão linear: ', reg.intercept_)
```

```python
y_pred = reg.predict(X_test)
reg.score(X_test, y_test)
print('Coeficiente de determinação do R²: ', reg.sc)
```

```python
print('R²:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
p = 1
while p<= 10:
    i = 0
    n = 100
    l =[]
    l2 = []
    while i <= n:
        X_num1 = excel_only_jogos[["price", "n_cat"]]
        X_train, X_test, y_train, y_test = train_test_split(X_num1, y, test_size=0.2)

        pol_reg = PolynomialFeatures(degree=p) #determinando quantos graus nossa função terá
        x_train =  pol_reg.fit_transform(X_train)
        x_test =  pol_reg.fit_transform(X_test)
        model = LinearRegression().fit(x_train, y_train) #criando novo modelo a partir dos dados transformados anteriormente
        y_pred = model.predict(x_test)
        k = model.score(x_test, y_test)
        m = np.sqrt(metrics.mean_squared_error(y_test, y_pred))
        l.append(m)
        l2.append(k)
        i += 1
    k = np.mean(l2)
    m = np.mean(l)
    print("Grau:",p,"R²:",m)
    print("Grau:",p,"Score",k)
    p+=1
```

Nesta regressão polinomial apresentamos para alguns graus de função e, como observado, também não foi possível obter resultados significantes, além de obter R² maiores que 1 quando o grau é maior que 5.

# Definição de Regressão linear
A regressão linear tem como base, criar uma reta que tenta minimizar a distância entre os diversos pontos de uma tabela, a fim de encontrar uma "Lei da Função" para aquele gráfico ou problema, para isso, utiliza-se os métodos dos mínimos quadrados. Dessa forma, a reta pode não ser perfeita (não passando por todos os pontos da tabela), mas mesmo assim, ser uma reta que tente minimizar a distância entre esses mesmos pontos. Para tratar dessa distância, utiliza-se o R², que vai de 0 a 1, demonstrando a "qualidade" da reta, onde "1" significa que a reta passa por todos os pontos.
No nosso caso, por exemplo, nosso R² foi de aproximadamente 0.18, demonstrando que a nossa reta estava com distâncias muito grandes entre os pontos, demonstrando a baixa relação entre o eixo X (sendo ele o preço, Idade, número de categorias, entre outras) e o eixo y (a porcentagem de positividade das avaliações).
