## Número de categorias

```python
n_cat = []
for i in steam.categories:
#     print(i.count(";")+1)
    n_cat.append(i.count(";")+1)
steam["n_cat"] = pd.Series(n_cat)
steam["n_cat"]
```

## Df só com jogos

```python
generos = steam.genres.str.cat(sep=";")
generos = generos.split(";")
lista_generos = []
for i in generos:
    if i not in lista_generos:
        lista_generos.append(i)
lista_software = ["Animation & Modeling", "Web Publishing", "Design & Illustration", "Early Access", "Utilities", "Video Production", "Nudity", "Photo Editing", "Accounting", "Game Development", "Free to Play", "Audio Production", "Tutorial", "Sexual Content", "Education", "Software Training", "Documentary"]
for i in lista_software:
    if i in lista_generos:
        lista_generos.remove(i)
# lista_generos
for i in lista_software:
    steam = steam[~steam.genres.str.contains(i)]
steam
```

## Eliminando jogos com tempo médio = 0

```python
steam = steam.loc[steam.average_playtime != 0]
steam
```

## Porcentagem positiva e classificação em True ou False

```python
steam["ratings"] = steam["positive_ratings"] + steam["negative_ratings"]
```

```python
steam["per_rating"] = steam["positive_ratings"]/steam["ratings"]
steam
```

```python
steam["positive"] = [True if x >=0.50 else False for x in steam["per_rating"]] 
steam.positive.value_counts()
```

# Efetuando o teste

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X_num1, y_ratings)

model = RandomForestClassifier(random_state = 0)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print(accuracy_score(y_test, y_pred))
```

# Média da acurácia de 10 testes

```python
accuracy = []
for i in range(10):
    X_train, X_test, y_train, y_test = train_test_split(X_num1, y_ratings)
    model = RandomForestClassifier(random_state = 0)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy.append(accuracy_score(y_test, y_pred))
print("acurácia:", np.mean(accuracy))
print("desvio:", np.std(accuracy))
```
